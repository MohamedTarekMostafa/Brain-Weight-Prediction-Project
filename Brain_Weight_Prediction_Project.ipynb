```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math

```

```python
df = pd.read_csv("headbrain.csv")
df.head()
```

```python
x = df['Head Size(cm^3)']
y = df['Brain Weight(grams)']
df.shape
```

```python
plt.scatter(x,y)
plt.xlabel('size (cm^3)')
plt.ylabel('Brain (grams)')
plt.title('Brain weight by size of head')
plt.show()

```

```python

x = np.array(x).reshape(-1,1)

y = np.array(y).reshape(-1,1)
```

```python
model  = LinearRegression()
model.fit(x,y)
```

```python

regression_model_mse = mean_squared_error(x,y)
print('Mean Squared Error:\t',math.sqrt(regression_model_mse))
print("R squared value\t\t",model.score(x,y))
```

```python

print(model.coef_[0])
print(model.intercept_[0])
```

```python
predict = model.predict(x)
```

```python
plt.scatter(x,y,color='blue')
plt.scatter(x,predict,color='red')
plt.title('Linear Regression')
plt.xlabel('Head Size')
plt.ylabel('Brain Weight')
```

```python
print('Prediction by the model:\t', model.predict([[2000]]))
```
